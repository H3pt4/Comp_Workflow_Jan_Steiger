1. What is nf-core?

nf-core is a collection of pipelines for working with (mainly) biological data

2. How many pipelines are there currently in nf-core?

112 in total, 66 released

3. Are there any non-bioinformatic pipelines in nf-core?

Generally, most of the pipelines in nf-core are bioinformatics pipelines, but there are some exceptions.

4. Let's go back a couple of steps. What is a pipeline and what do we use it for?

A pipeline is a tool for combining several processes to streamline a workflow, which we use to for processing data with different tools to generate reproducible output for a modified input

5. Why do you think nf-core adheres to strict guidelines?

Strict guidelines allow other users to understand the documentation and usage of pipelines more easily, since they give an easy-to-understand frame of reference and they make it reproducible.

6. What are the main features of nf-core pipelines?

Extensive documentation, stable releases, open source, CI testing, portability/reproducability, packaging
